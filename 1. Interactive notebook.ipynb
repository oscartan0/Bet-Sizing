```python
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Sliders based on absolute bet size
win_prob_slider = widgets.FloatSlider(value=0.55, min=0.4, max=0.8, step=0.01, description="Win Probability")
num_bets_slider = widgets.IntSlider(value=100, min=1, max=1000, step=1, description="# of Bets")
num_simulations_slider = widgets.IntSlider(value=1000, min=10, max=10000, step=10, description="# of Simulations")
bet_size_slider = widgets.FloatSlider(value=1, min=0, max=100, step=0.1, description="Bet Size")

print("Initial money will be set at $100: please select all other parameters of the simulation")
# Display sliders
display(win_prob_slider, num_bets_slider, num_simulations_slider, bet_size_slider)
```

```python
win_prob = win_prob_slider.value
num_bets = int(num_bets_slider.value)
num_simulations = int(num_simulations_slider.value)
bet_size = bet_size_slider.value
starting_money = 100

# Pre-generate all wins/losses to increase speed
wins_matrix = np.random.rand(num_simulations, num_bets) < win_prob

# Create array for money to run simulations simultaneously
money = np.full((num_simulations,), starting_money, dtype=float)
ruin = np.zeros(num_simulations, dtype=bool)
# Drawdown Calculations
max_drawdown = np.zeros(num_simulations) # drawdown at least 0
peak = money.copy()
# Run simulations
for i in range(num_bets):
    bet = np.minimum(bet_size, money) #caps bet at amount you have
    money += np.where(wins_matrix[:, i], bet, -bet) #(comment)
    ruin |= (money <= 0) # ruin defined as run out of money
    peak = np.maximum(peak, money)
    drawdown = (peak - money) / peak
    max_drawdown = np.maximum(max_drawdown, drawdown)

# Drawdown and ruin (risk measures)
prob_ruin = np.mean(ruin)
avg_max_drawdown = np.mean(max_drawdown)

# Results
avg_payoff = np.mean(money)
avg_sqrt_utility = np.mean(np.sqrt(money))
avg_log_utility = np.mean(np.log(np.maximum(money,1e-100)))


returns= (money - starting_money)/starting_money

avg_return = np.mean(returns)
return_variance = np.var(returns)   
return_std_dev = np.std(returns)

print(f"average payoff is: {round(avg_payoff,3)}")
print(f"average sqrt utility is: {round(avg_sqrt_utility,3)}")
print(f"average log utility is: {round(avg_log_utility,3)}")
print(f"average return is: {round(avg_return*100,3)}%")
print(f"variance of returns is: {round(return_variance*100,3)}%")
print(f"standard deviation of returns: {round(return_std_dev*100,3)}%")
print(f"probability of ruin: {round(prob_ruin*100,2)}%")
print(f"average maximum drawdown: {round(avg_max_drawdown*100,2)}%")
```

```python
# Slider for bet percentage
bet_percentage_slider = widgets.FloatSlider(value=1, min=0, max=100, step=1, description="Bet Percentage")

print("Other parameters will be kept as above, please select bet percentage")
# Display sliders
display(bet_percentage_slider)
```

```python
win_prob = win_prob_slider.value
num_bets = int(num_bets_slider.value)
num_simulations = int(num_simulations_slider.value)
bet_percentage = bet_percentage_slider.value/100
starting_money = 100

# Pre-generate all wins/losses to increase speed
wins_matrix = np.random.rand(num_simulations, num_bets) < win_prob

# Create array for money to run simulations simultaneously
money = np.full((num_simulations,), starting_money, dtype=float)
ruin = np.zeros(num_simulations, dtype=bool)
# Drawdown Calculations
max_drawdown = np.zeros(num_simulations) # drawdown at least 0
peak = money.copy()
# Run simulations
for i in range(num_bets):
    bet = money * bet_percentage
    money += np.where(wins_matrix[:, i], bet, -bet) 
    ruin |= (money <= 1) # ruin defined as $1 as betting percentages should never lead to 0
    peak = np.maximum(peak, money)
    drawdown = (peak - money) / peak
    max_drawdown = np.maximum(max_drawdown, drawdown)

# Results
avg_payoff = np.mean(money)
avg_sqrt_utility = np.mean(np.sqrt(money))
avg_log_utility = np.mean(np.log(np.maximum(money, 1e-100))) #don't let money drop to 0

returns= (money - starting_money)/starting_money

avg_return = np.mean(returns)
return_variance = np.var(returns) 
return_std_dev = np.std(returns)

# Drawdown and ruin (risk measures)
prob_ruin = np.mean(ruin)
avg_max_drawdown = np.mean(max_drawdown)

print(f"the average payoff is {round(avg_payoff,3)}")
print(f"the average sqrt utility is {round(avg_sqrt_utility,3)}")
print(f"the average log utility is {round(avg_log_utility,3)}")
print(f"the average return is {round(avg_return*100,3)}%")
print(f"the variance of returns is {round(return_variance*100,3)}%")
print(f"the standard deviation of returns is {round(return_std_dev*100,3)}%")
print(f"probability of ruin: {round(prob_ruin*100,2)}%")
print(f"average maximum drawdown: {round(avg_max_drawdown*100,2)}%")
```
